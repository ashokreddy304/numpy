## What is the difference between list and numpy array

1. Similarities:
   - Both can be used to store data.
   - The order will be preserved in both. Hence, indexing and slicing concepts are applicable.
   - Both mutable. We can change its content.
     
2. Difference
   - A list is a Python built-in type. We have to install and import numpy explicitly.
   - A list can contain heterogeneous elements. But the array contains homogeneous elements.
   - On the list, we can't perform vector operations. But on ndarray, we can perform vector operations.
   - Arrays consume less memory than lists.
   - Arrays are super past when compared with lists.
   - NumPy arrays are more convenient to use while performing complex mathematical operations.

# View vs Copy

## view
- View is not a separate object and is just is logical representation of the existing array.
- If any changes are made to the original array, those changes will be reflected in the view. vice-versa also.
- We can create view explicitly by using view() method of ndarray class

```python
a = np.array([10,20,30,40])
print(a)

b = a.view()
print(b)
```

```python
a[0] = 200
print(a)
print(b)
```

 ## copy()
- Copy means a separate object.
- If we perform any changes to the original array, those changes won't be reflected in the copy. Vice-versa also.
- By using the copy() method of ndarray, we can create a copy of an existing ndarray.

```python
a = np.array([10,20,30,40])
print(a)

b = a.copy()
print(b)
```

```python
a[0] = 200
print(a)
print(b)
```

### Slicing - A Group of elements that are in order.
- syntax-1 l[begin:end:step]
- begin, end, and step values can be either +ve or -ve. These are optional.
- If begin is not specified, then zero (0) will be taken by default.
- If the end is not specified, then up to the end of the list will be taken.
- If the step is not specified, then 1 will be taken by default.
- If the step value is +ve, then in the forward direction, begin to end-1 will be considered.
- If the step value is -ve, then in the backward direction, begin to end+1 will be considered.
- In the forward direction, if the end is 0, then the result will always be empty.
- In the backward direction, if the end is -1, then the result will always be empty.

### Stacking:
- Horizontal stack: np.hstack()
- Vertical stack: np.vstack()
- Depth stack: np.dstack()

#### Horizontal stack: np.hstack()
- Purpose: Stacks arrays side-by-side (columns increase).
- Requires arrays with the same number of rows.

```python
a = np.array([[1,2,3],[4,5,6]])
print(a.shape)
b = np.array([[7,8,9],[10,11,12]])
print(b.shape)
```

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8,9],[10,11,12]])
result = np.hstack((a,b))
result
```

#### Vertical Stack  np.vstack()
- Purpose: Stacks arrays top-to-bottom (rows increase).
- Requires arrays with the same number of columns.

```python
a = np.arange(10)
a1 = a.reshape(5,2)
a1
```

```python
b = np.arange(10,20)
b1 = b.reshape(5,2)
b1
```

```python
final_vec = np.vstack((a1,b1))
final_vec
```

#### Depth Stack  np.dstack()
- Purpose: Stacks arrays along the third axis (depth increases).
- Think of it like stacking pages in a book.
- Requires arrays with the same shape

```python
result = np.dstack((a1, b1))
result
```

```python
result.shape
```

## Broadcasting

- Broadcasting in NumPy is a feature that allows NumPy to perform arithmetic operations on arrays of different shapes without making unnecessary copies of data.
- Instead of manually repeating (replicating) elements to match shapes, NumPy automatically expands dimensions so operations work element-wise.

### Key Rules of Broadcasting
- When operating on two arrays, NumPy compares shapes from right to left:
- If dimensions are equal   compatible.
- If one of them is 1   can be stretched to match the other.
- If neither is 1 and dimensions differ   error.

```python
a = np.array([1, 2, 3])
print(a + 5)  # Here, 5 is broadcast to [5, 5, 5]
```

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([10, 20, 30])
A + B
```

```python
A = np.array([[1], [2], [3]])  # shape (3,1)
B = np.array([10, 20, 30])     # shape (3,)
A + B
```
